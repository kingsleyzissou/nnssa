```python
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler, SMOTE
from numpy import where
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
    n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
```

```python
counter = Counter(y)
```

```python
plt.figure(figsize=(15,10))
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
under = RandomUnderSampler()
X_under, y_under = under.fit_resample(X, y)
counter_under = Counter(y_under)
```

```python
plt.figure(figsize=(15,10))
for label, _ in counter_under.items():
    row_ix = where(y_under == label)[0]
    plt.scatter(X_under[row_ix, 0], X_under[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
over = RandomOverSampler(sampling_strategy=0.6)
X_over, y_over = over.fit_resample(X, y)
counter_over = Counter(y_over)
print(counter_over)
```

```python
plt.figure(figsize=(15,10))
for label, _ in counter_over.items():
    row_ix = where(y_over == label)[0]
    plt.scatter(X_over[row_ix, 0], X_over[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X, y)
counter_smote = Counter(y_smote)
```

```python
plt.figure(figsize=(15,10))
for label, _ in counter_smote.items():
    row_ix = where(y_smote == label)[0]
    plt.scatter(X_smote[row_ix, 0], X_smote[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```
